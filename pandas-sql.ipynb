```python
# payment table
# 월별 전체 매출, 평균매출, 결제갯수
# 위의 값을 출력하고 위의 각각의 값을 
# dataframe 의 칼럼값으로 해서 출력
# pandas, sql 
```

```python
sql = '''
select * from payment;
'''
pay = pd.read_sql(sql,db)
```

```python
pay.head(2)
```

```python
pay['month'] = pay['payment_date'].\
apply(lambda x: str(x)[:7])
```

```python
pay.head(3)
```

```python
paysum = pay.groupby('month')['amount'].sum()
paymean = pay.groupby('month')['amount'].mean()
paycount = pay.groupby('month')['amount'].count()
print(paysum)
print(paymean)
print(paycount)
```

```python
import numpy as np
pay.groupby('month').\
agg({'amount' : [np.mean, np.sum],'payment_id' : np.size})
```

```python
pay2 = pay[['payment_id', 'amount','payment_date']]
pay2['month'] = pay['payment_date'].apply(lambda x: str(x)[:7])
pay2
```

```python
sql = '''
select payment_id, left(payment_date,7) 'month',avg(amount) 평균매출 ,
sum(amount) 총매출, count(payment_id) size
from payment
group by month;
'''
pd.read_sql(sql,db)
```
